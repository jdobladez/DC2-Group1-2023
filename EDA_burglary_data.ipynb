## Exploring time trends

Check mean & std??

```python
for index, value in enumerate(df['Month']):
    df.at[index, 'Month'] = datetime.strptime(value, '%Y-%m')
    
df.head()
```

```python
df_per_month = df.groupby(pd.Grouper(key='Month', freq='M')).size()
print(df_per_month)

df_per_month.plot(kind='line', marker='o', figsize=(10, 6), c='purple')
plt.xlabel('Month')
plt.ylabel('Number of Data Points')
plt.title('Data Points per Month')
plt.show()
```

```python
sns.heatmap(df_per_month.to_frame(), cmap='YlGnBu')
plt.title('Number of records per month')
plt.show()

df_per_month.plot(kind='bar', figsize=(10,5))
plt.title('Number of records per month')
plt.xlabel('Month')
plt.ylabel('Number of records')
plt.show()
```

```python
highest_per_year = df_per_month.groupby(df_per_month.index.year).idxmax()
print(highest_per_year)
```

```python
lowest_per_year = df_per_month.groupby(df_per_month.index.year).idxmin()
print(lowest_per_year)
```

```python
df_per_year = df.groupby(pd.Grouper(key='Month', freq='Y')).size()
print(df_per_year)
```

```python
plt.bar(df_per_year.index.year, df_per_year.values)
plt.title('Number of Records per Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

## Last outcome category

```python
df_grouped_by_outcome = df.groupby(pd.Grouper(key='Last outcome category')).size()

df_grouped_by_outcome.plot.bar()
plt.title('Counts by Outcome Category')
plt.xlabel('Outcome Category')
plt.ylabel('Counts')
plt.show()
```

```python
df_per_month = pd.DataFrame(df_per_month)
df_per_month['Year'] = df_per_month.index.year
df_per_month['M'] = df_per_month.index.month
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

for year in df_per_month['Year'].unique():
    year_data = df_per_month[df_per_month['Year'] == year]
    ax.plot(year_data['M'], year_data[0], label=str(year))

ax.set_title('Burglaries per Month by Year')
ax.set_xlabel('Month')
ax.set_ylabel('Count')
ax.legend()
plt.show()
```
