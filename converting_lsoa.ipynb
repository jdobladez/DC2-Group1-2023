```python
import pandas as pd
import numpy as np
```

```python
excel_file = "Lower_Layer_Super_Output_Area_(2011)_to_Ward_(2015)_Lookup_in_England_and_Wales.csv"
df_temp = pd.read_csv(excel_file)
df_temp
```

```python
df = df_temp.iloc[:, [0, 1, 2]]
df = df.rename(columns={'LSOA11CD': 'LSOA code'}, inplace=False)
df = df.rename(columns={"WD15CD" : "New code"}, inplace=False)
df
```

```python
df_2 = pd.read_csv("ward_median_house_merged_cleaned.csv")
df_2
```

```python
merged_df = pd.merge(df_2, df, on='New code', how='left')
merged_df
```

```python
df_data = pd.read_csv("burglary_data.csv")
df_data
```

```python
merged_df_2 = pd.merge(merged_df, df_data, on='LSOA code', how='left')
merged_df_2
```

```python
merged_df_2.to_csv("LSOA_name_ward_median_house.csv", index=False)
```
